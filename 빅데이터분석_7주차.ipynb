```python
%config InlineBackend.figure_format = 'retina'
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

sns.barplot(traffic)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(traffic[['사망자수(명)']].T)
plt.show()
```
