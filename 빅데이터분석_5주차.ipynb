```python
import matplotlib.pyplot as plt
# 전체 제목 — 어느 위치에서 호출해도 무관 (Figure 레벨)
plt.suptitle('전체 제목', fontsize=14, y=1.02)
# subplot 1 — 생성 직후 바로 title 지정
plt.subplot(2, 3, 1)
plt.plot([1, 4, 9, 16])
plt.title('제곱수')
# subplot 5
plt.subplot(2, 3, 6)
plt.plot([1, 2, 3, 4])
plt.title('선형')
# subplot 2
plt.subplot(2, 3, 2)
plt.plot([6, 5, 2, 1])
plt.title('감소')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, 3, figsize=(7, 4))
fig.set_size_inches(6, 3)
ax[0, 1].plot([1, 4, 9, 16])
ax[1, 0].plot([1, 2, 3, 4])
ax[1, 2].plot([6, 5, 2, 1])
# 전체 제목 — Figure 레벨
fig.suptitle('전체 제목', fontsize=14, fontweight='bold')
# 각 subplot 제목 — Axes 레벨
ax[0, 0].set_title('(0,0) 빈 플롯')
ax[0, 1].set_title('(0,1) 제곱수')
ax[0, 2].set_title('(0,2) 빈 플롯')
ax[1, 0].set_title('(1,0) 선형')
ax[1, 1].set_title('(1,1) 빈 플롯')
ax[1, 2].set_title('(1,2) 감소')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(6, 4))
ax1 = fig.add_subplot(231)
ax1.plot([1, 4, 9, 16])
ax1.set_title('제곱수')
ax2 = fig.add_subplot(233)
ax2.plot([1, 2, 3, 4])
ax2.set_title('선형')
ax3 = fig.add_subplot(235)
ax3.plot([6, 5, 2, 1])
ax3.set_title('감소')
fig.suptitle('전체 제목', fontsize=14)
plt.tight_layout()
plt.show()
```
