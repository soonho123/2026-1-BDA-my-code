```python
value = 'apple'
match value:
  case 'apple':
    result = '사과'
  case ' banana':
    result = '바나나'

print(result)

```

```python
value = 'banana'
match value:
  case 'apple':
    result = "사과"
  case ' banana':
    result = "바나나"

print(result)
```
